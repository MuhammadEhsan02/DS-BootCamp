```python
from bokeh.plotting import ColumnDataSource, figure, output_file, show

output_file("demo.html")

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y=[2, 5, 8, 2, 7],
    desc=['A', 'b', 'C', 'd', 'E'],
))

TOOLTIPS = [
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("desc", "@desc"),
]

p = figure(width=400, height=400, tooltips=TOOLTIPS,
           title="Mouse over the dots")

p.circle('x', 'y', size=20, source=source)

show(p)
```

```python
import pandas as pd

from bokeh.palettes import Spectral4
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT

p = figure(width=800, height=250, x_axis_type="datetime")
p.title.text = 'Click on legend entries \
            to hide the corresponding lines'

for data, name, color in zip([AAPL, IBM, MSFT,
                              GOOG], ["AAPL", "IBM",
                                      "MSFT", "GOOG"],
                             Spectral4):
    df = pd.DataFrame(data)
    df['date'] = pd.to_datetime(df['date'])
    p.line(df['date'], df['close'], line_width=2,
           color=color, alpha=0.8,
                           legend_label=name)

p.legend.location = "top_left"
p.legend.click_policy = "hide"

output_file("interactive_legend.html",
            title="interactive_legend.py example")

show(p)
```

```python
import pandas as pd
from bokeh.palettes import Spectral4
from bokeh.plotting import figure, output_file, show
from bokeh.sampledata.stocks import AAPL, GOOG, IBM, MSFT
from bokeh.models import HoverTool, NumeralTickFormatter, ColumnDataSource, CustomJS
from bokeh.models.widgets import Slider
from bokeh.layouts import column

# Define the output file
output_file("interactive_legend_slider.html", title="Interactive Legend with Slider")

# Create a figure
p = figure(width=800, height=400, x_axis_type="datetime")

# Title and labels
p.title.text = "Stock Prices Over Time"
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Price"

# Create a HoverTool for interactive tooltips
hover = HoverTool()
hover.tooltips = [("Date", "@date{%F}"), ("Price", "@close{$0.2f}")]
hover.formatters = {"@date": "datetime"}
p.add_tools(hover)

# Define line colors and legend labels
colors = Spectral4
names = ["AAPL", "IBM", "MSFT", "GOOG"]

# Create a data source for the slider callback
source = ColumnDataSource(data=dict(x=[], y=[]))

# Create a Slider widget
slider = Slider(start=10, end=len(AAPL), value=len(AAPL), step=10, title="Number of Data Points")

# Callback to update the data source based on slider value
callback = CustomJS(args=dict(source=source, slider=slider), code="""
    var data = source.data;
    var n = slider.value;
    data['x'] = data['x'].slice(0, n);
    data['y'] = data['y'].slice(0, n);
    source.change.emit();
""")

slider.js_on_change('value', callback)

# Plot the data and customize the lines
for i, (data, name, color) in enumerate(zip([AAPL, IBM, MSFT, GOOG], names, colors)):
    df = pd.DataFrame(data)
    df['date'] = pd.to_datetime(df['date'])
    source.data = dict(x=df['date'], y=df['close'])
    p.line('x', 'y', source=source, line_width=2, color=color, alpha=0.8, legend_label=name)

# Add a legend with hide/show functionality
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Customize axis formatting
p.xaxis.formatter = DatetimeTickFormatter(days="%b %d")
p.yaxis.formatter = NumeralTickFormatter(format="$0.00")

# Create a layout with the plot and slider
layout = column(p, slider)

# Show the plot
show(layout)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool, ColumnDataSource, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import random

# Define the output file
output_file("interactive_scatter_plot.html", title="Interactive Scatter Plot Example")

# Create random data for the scatter plot
data = {
    'x': [random.random() for _ in range(100)],
    'y': [random.random() for _ in range(100)],
    'size': [random.randint(5, 20) for _ in range(100)],
    'color': [random.randint(0, 255) for _ in range(100)]
}

# Create a ColumnDataSource for the scatter plot
source = ColumnDataSource(data=data)

# Create a linear color mapper based on the 'color' field
color_mapper = linear_cmap(field_name='color', palette=Viridis256, low=min(data['color']), high=max(data['color']))

# Create a figure
p = figure(width=800, height=400, tools="pan,box_zoom,wheel_zoom,reset,save")

# Add a scatter plot
scatter = p.scatter('x', 'y', source=source, size='size', color=color_mapper, legend_field='color', fill_alpha=0.6)

# Create a HoverTool for interactive tooltips
hover = HoverTool()
hover.tooltips = [("X", "@x"), ("Y", "@y"), ("Size", "@size"), ("Color", "@color")]
p.add_tools(hover)

# Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Customize plot appearance
p.legend.title = "Color"
p.legend.label_text_font_size = "10pt"

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.transform import dodge
from bokeh.models import HoverTool
import random

# Define the output file
output_file("interactive_bar_chart.html", title="Interactive Bar Chart Example")

# Sample data for a hypothetical use case
categories = ['Category A', 'Category B', 'Category C', 'Category D']
years = [2020, 2021, 2022, 2023]
data = {'Categories': categories}

for year in years:
    data[str(year)] = [random.randint(10, 100) for _ in categories]

# Create a ColumnDataSource for the bar chart
source = ColumnDataSource(data=data)

# Create a figure
p = figure(x_range=categories, width=800, height=400, title="Yearly Data by Category",
           toolbar_location=None, tools="")

# Create stacked bar glyphs
p.vbar(x=dodge('Categories', -0.25, range=p.x_range), top='2020', width=0.2, source=source, color="#c9d9d3", legend_label="2020")
p.vbar(x=dodge('Categories',  0.0,  range=p.x_range), top='2021', width=0.2, source=source, color="#718dbf", legend_label="2021")
p.vbar(x=dodge('Categories',  0.25, range=p.x_range), top='2022', width=0.2, source=source, color="#e84d60", legend_label="2022")
p.vbar(x=dodge('Categories',  0.5,  range=p.x_range), top='2023', width=0.2, source=source, color="#ffbc79", legend_label="2023")

# Add a hover tool for interactive tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Categories"), ("2020", "@2020"), ("2021", "@2021"), ("2022", "@2022"), ("2023", "@2023")]
p.add_tools(hover)

# Customize the plot appearance
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool
import random

# Define the output file
output_file("interactive_line_chart.html", title="Interactive Line Chart Example")

# Sample data for a use case
years = list(range(2010, 2021))
sales = [random.randint(500, 1000) for _ in years]
expenses = [random.randint(300, 800) for _ in years]

# Create a figure
p = figure(width=800, height=400, title="Sales and Expenses Over Time", x_axis_label="Year", y_axis_label="Amount ($)")

# Plot sales data as a line
p.line(years, sales, line_width=2, line_color="blue", legend_label="Sales")

# Plot expenses data as a line
p.line(years, expenses, line_width=2, line_color="red", legend_label="Expenses")

# Add a HoverTool for interactive tooltips
hover = HoverTool()
hover.tooltips = [("Year", "@x"), ("Amount", "@y{$0,0}")]
p.add_tools(hover)

# Customize the plot appearance
p.legend.location = "top_left"

# Show the plot
show(p)
```
