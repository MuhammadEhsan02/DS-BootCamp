```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df
```

```python
df.info()
```

```python
df.describe
```

```python
plt.scatter(df['Annual_Income_(k$)'], df['Spending_Score'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.title('Annual Income vs Spending Score')
plt.show()
```

```python
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
gender_counts = df['Genre'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.boxplot(df['Spending_Score'])
plt.ylabel('Spending Score')
plt.title('Spending Score Distribution')
plt.show()
```

```python
plt.plot(df['CustomerID'], df['Spending_Score'])
plt.xlabel('CustomerID')
plt.ylabel('Spending Score')
plt.title('Spending Score by CustomerID')
plt.show()
```

```python
genre_counts = df['Genre'].value_counts()
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.violinplot(df['Age'], vert=False)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
pd.plotting.scatter_matrix(df[['Age', 'Annual_Income_(k$)', 'Spending_Score']], figsize=(8, 8))
plt.suptitle('Scatter Matrix Plot')
plt.show()
```

```python
income_groups = df.groupby(pd.cut(df['Annual_Income_(k$)'], bins=[0, 30, 60, 90, 120, 150])).size()
plt.barh(income_groups.index.astype(str), income_groups.values)
plt.xlabel('Count')
plt.ylabel('Annual Income (k$)')
plt.title('Annual Income Groups')
plt.show()
```

```python
age_groups = df.groupby(pd.cut(df['Age'], bins=[0, 20, 30, 40, 50, 60, 70, 80])).size()
plt.stackplot(age_groups.index.astype(str), age_groups.values, labels=age_groups.index.astype(str), alpha=0.7)
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.legend()
plt.show()
```
