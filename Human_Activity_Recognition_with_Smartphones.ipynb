# Data Processing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

x = df.drop('Activity', axis=1)
y = df['Activity']
```

```python
x
```

```python
y
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
random_state = 42
test_size = 0.25
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=test_size, random_state=random_state)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred_knn)
print('KNN Model Accuracy:', acc)
```

# Naive Bayes

**Bernoulli**

```python
from sklearn.naive_bayes import BernoulliNB

nb_bernoulli = BernoulliNB()
nb_bernoulli.fit(x_train, y_train)
```

```python
y_pred_nb_bernoulli = nb_bernoulli.predict(x_test)
```

```python
acc_bernoulli = accuracy_score(y_test, y_pred_nb_bernoulli)
print('Bernoulli Naive Bayes Accuracy:', acc_bernoulli)
```

**Multinomial**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
nb_multinomial = MultinomialNB()
nb_multinomial.fit(x_train_scaled, y_train)
```

```python
y_pred_nb_multinomial = nb_multinomial.predict(x_test_scaled)
```

```python
acc_multinomial = accuracy_score(y_test, y_pred_nb_multinomial)
print('Multinomial Naive Bayes Accuracy:', acc_multinomial)
```

# SVM Model

```python
from sklearn.svm import SVC

svm = SVC(random_state=random_state)
svm.fit(x_train, y_train)
```

```python
y_pred_svm = svm.predict(x_test)
```

```python
acc_svm = accuracy_score(y_test, y_pred_svm)
print('SVM Model Accuracy:', acc_svm)
```
